## SPLITING X & Y VALUES

```python
x=df[features]
y=df['strength']
```

## TRAIN-TEST SPLIT

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## SCALING

```python
scaler=StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

## MODELS IMPLEMENTED

## LINEAR REGRESSION

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
print('LinearRegression R2:',r2_score(y_test,y_pred))
print('LinearRegression MSE:',mean_squared_error(y_test,y_pred))
```

## DECISION TREE

```python
dt = DecisionTreeRegressor(random_state=42)
dt.fit(x_train,y_train)
```

```python
y_pred_dt = dt.predict(x_test)
print('Decisiontree R2:',r2_score(y_test,y_pred_dt))
print('Decisiontree MSE:',mean_squared_error(y_test,y_pred_dt))
```

## RANDOM FOREST

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
print('Randomforest R2:',r2_score(y_test,y_pred_rf))
print('Randomforest MSE:',mean_squared_error(y_test,y_pred_rf))
```

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor(random_state=42, n_jobs=-1)
xgb.fit(x_train,y_train)
```

```python
y_pred=xgb.predict(x_test)
print('xgb r2:',r2_score(y_test,y_pred))
print('xgb MSE:',mean_squared_error(y_test,y_pred))
```

## FINAL RESULTS

```python

comparison_df = pd.DataFrame({
    "Model": ["Linear Regression", "Decision Tree", "Random Forest", "XGBoost"],
    "R2 Score": [
        r2_score(y_test,lr.predict(x_test)),
        r2_score(y_test,dt.predict(x_test)),
        r2_score(y_test,rf.predict(x_test)),
        r2_score(y_test,xgb.predict(x_test))
    ],
    "MSE": [
        mean_squared_error(y_test,lr.predict(x_test)),
        mean_squared_error(y_test,dt.predict(x_test)),
        mean_squared_error(y_test,rf.predict(x_test)),
        mean_squared_error(y_test,xgb.predict(x_test))
    ]
})

comparison_df
```

## HYPERPARAMETER TUNING

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100,200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

grid_xgb = GridSearchCV(xgb, param_grid, cv=5, scoring='r2')
grid_xgb_mse = GridSearchCV(xgb, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_xgb.fit(x_train, y_train)
grid_xgb_mse.fit(x_train,y_train)
```

```python
print("Best XGBoost Params:", grid_xgb.best_params_)
print("Best CV R2:", grid_xgb.best_score_)
print(" Best CV MSE:", -grid_xgb_mse.best_score_)
```
